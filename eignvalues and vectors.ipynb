```python
import numpy as np
from scipy.sparse import random
from scipy.sparse.linalg import eigsh

# Generate a large sparse symmetric matrix
size = 1000  # Size of the matrix
density = 0.01  # Fraction of non-zero elements

# Generate a random sparse matrix
sparse_matrix = random(size, size, density=density, format='csr')
symmetric_sparse_matrix = (sparse_matrix + sparse_matrix.T) / 2

# Compute the top 5 eigenvalues and their corresponding eigenvectors
k = 5  # Number of eigenvalues/vectors to compute
eigenvalues, eigenvectors = eigsh(symmetric_sparse_matrix, k=k)

# Display results
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```
